```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
##imputer.fit(coal)
```

```python
m = imputer.fit_transform(coal)
```

```python
imputer.statistics_
```

```python
coal_tr = pd.DataFrame(m, columns=coal.columns)
```

```python
coal_tr.describe()
```

# Creating Design

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([('imputer',SimpleImputer(strategy="median")),('std_scaler', StandardScaler())])
```

```python
coal_num_tr = my_pipeline.fit_transform(coal_tr)
```

```python
coal_num_tr
```

# Selection of suitable model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
##model = LinearRegression()
##model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(coal_num_tr, coal_labels)
```

```python
some_data = coal.iloc[:5]
```

```python
some_labels = coal_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
coal_predictions = model.predict(coal_num_tr)
lin_mse = mean_squared_error(coal_labels, coal_predictions)
lin_rmse = np.sqrt(lin_mse)
```

```python
lin_mse
```

```python
lin_rmse
```

Cross Validation...as we saw above our data has been overfitted

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,coal_num_tr,coal_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores" , scores)
    print("Mean: ",scores.mean())
    print("Standard deviation: ",scores.std())
```

```python
print_scores(rmse_scores)
```

# Saving the model

```python
from joblib import dump,load
dump(model, 'Coal.joblib')
```

# Testing the model on test set

```python
X_test = test_set.drop("Gross Calorific Value" , axis=1)
Y_test = test_set["Gross Calorific Value"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```
